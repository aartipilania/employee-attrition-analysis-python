```python
#Finding the attrition rate of this company
df['Attrition'].value_counts(normalize=True)['Yes']*100 
```

```python
df['Attrition'].value_counts()
##The company experienced an overall attrition rate of 16.1%, with 237 employees leaving out of a workforce of 1,470.
```

```python
##Department wise attrition wise analysis
dept_attr=pd.crosstab(
    df['Department'],
    df['Attrition'],
    normalize='index'
) * 100
dept_attr.round(2)
##The Sales department experienced the highest attrition rate (20.63%), followed by Human Resources (19.05%).
##Research & Development showed the lowest attrition rate (13.84%), indicating better employee retention.
```

```python
##Visualising the above findings

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##Visualising the above findings
dept_attr['Yes'].sort_values().plot(kind='barh')
plt.title('Attrition Rate by Department')
plt.xlabel('Attrition Rate (%)')
plt.savefig("Departmentwiseattrition.png",
            bbox_inches='tight',
            dpi=300)
plt.show()
```

```python
##Do employees who work overtime leave more frequently?
overtime_attr=pd.crosstab(df["OverTime"],df['Attrition'],normalize='index')*100
overtime_attr.round(2)
```

```python
##Visualising the above findings
overtime_attr["Yes"].sort_values().plot(kind='bar')
plt.title("Overtime as a contributor to Attrition")
plt.xlabel("Overtime") 
plt.ylabel("Attrition Rate (%)")
plt.savefig("Attritionovertime.png",
            bbox_inches='tight',
            dpi=300)
plt.show()
##Employees who work overtime have an attrition rate of 30.53%, nearly three times that of employees who do not 
##work overtime (10.44%). This suggests overtime may be a significant driver of employee turnove
```

```python
##which job role has highest attrition?
jobrole_attr=pd.crosstab(
    df['JobRole'],
    df['Attrition'],
    normalize='index'
) * 100
jobrole_attr.round(2)
```

```python
##Visualising the above finding
jobrole_attr["Yes"].sort_values().plot(kind='barh')
plt.title("Attrition by Job Roles")
plt.ylabel("Job Role")
plt.xlabel("Attrition Rate %")
plt.savefig("Attritionbyjobrole.png",
            bbox_inches='tight',
            dpi=300)
plt.show()
```

```python
#Findings
##Sales Representatives exhibit the highest attrition rate (39.76%), significantly higher than all other job roles. 
##Human Resources and Laboratory Technicians also show relatively elevated attrition rates above 20%. In contrast, managerial and director-level roles experience very low attrition, suggesting greater employee retention in senior positions
```

```python
#Attrition on the basis of income 
##Creating Income bands for easier analysis
df['Income_Band'] = pd.cut(
    df['MonthlyIncome'],
    bins=[0,4000,8000,12000,16000,20000],
    labels=['<4K','4K-8K','8K-12K','12K-16K','16K-20K']
)
df['Income_Band']
```

```python
##Attrition rate on the basis of Income band
income_attr=pd.crosstab(df['Income_Band'],df["Attrition"], normalize='index')*100
income_attr["Yes"]
```

```python
income_attr["Yes"].sort_values().plot(kind='barh')
plt.title("Attrition by Income")
plt.ylabel("Income Band")
plt.xlabel("Attrition Rate %")
plt.savefig("Attritionbyincome.png",
            bbox_inches='tight',
            dpi=300)
plt.show()
```

```python
sns.boxplot(x="Attrition", y="YearsAtCompany", data=df)
plt.savefig("Attritionbyyearsatcompany.png",
            bbox_inches='tight',
            dpi=300)
plt.show()
```

```python
#Correlation Heatmap
df["Attrition_num"] = df["Attrition"].map({"No":0,"Yes":1})
cols = [
    "Age",
    "MonthlyIncome",
    "YearsAtCompany",
    "TotalWorkingYears",
    "DistanceFromHome",
    "Attrition_num"
]

corr = df[cols].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.savefig("correlation_heatmap.png",
            bbox_inches='tight',
            dpi=300)
plt.show()


```
